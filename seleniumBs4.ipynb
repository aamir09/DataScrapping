# Selenium With BeautifulSoup Tutorial[Python]

```python
from selenium import webdriver  
import time  
from bs4 import  BeautifulSoup
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import pandas as pd
import numpy as np
```

```python

path="C:/Users/Aamir/Downloads/chromedriver_win32/chromedriver.exe"
service=Service(path)
browser = webdriver.Chrome(service=service)
browser.get('https://1xbet.whoscored.com')

## Finding The Search-Box ## 
sbox=browser.find_element('id','search-box')

## Passing Values to the search Box ##
sbox.clear()
sbox.send_keys('Manchester United')

## Submitting the Search ##
sbutton=browser.find_element('id','search-button')
sbutton.click()

## Find first result ##

result=browser.find_element_by_class_name('search-result')
link=result.find_element_by_tag_name('a') 
url=link.get_property('href')
print(url)

#Go to the first results link 
browser.get(url)
#Get the html page
html=browser.page_source

browser.close()






```

```python
#Parse the data and store into a variable 
soup=BeautifulSoup(html,'html.parser')
type(soup)
```

```python
#Create Dataset
data=[]
names=soup.find_all('span',{'class':'iconize iconize-icon-left'})[1:]
ratings=soup.find_all('td',{'class':'rating sorted'})
for i,j in zip(names,ratings):
    datum={}
    datum['PlayerName']=i.text
    datum['PlayerRatings']=j.text
    data.append(datum)

dataset=pd.DataFrame(data)
dataset.head(24)
    
```
